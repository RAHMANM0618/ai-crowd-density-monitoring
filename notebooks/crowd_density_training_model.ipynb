```python
import os, glob, numpy as np, cv2, scipy.io as sio, torch
from torch.utils.data import Dataset, DataLoader

def gaussian_kernel(size=15, sigma=4):
    ax = np.arange(-size//2 + 1., size//2 + 1.)
    xx, yy = np.meshgrid(ax, ax)
    k = np.exp(-(xx**2 + yy**2) / (2. * sigma**2))
    return (k / k.sum()).astype(np.float32)

KERNEL = gaussian_kernel(15, 4)

def make_density_map(img_h, img_w, points):
    dm = np.zeros((img_h, img_w), dtype=np.float32)
    for x,y in points:
        xi, yi = min(int(x), img_w-1), min(int(y), img_h-1)
        x1, y1 = xi-7, yi-7
        x2, y2 = xi+8, yi+8
        kx1, ky1 = max(0,-x1), max(0,-y1)
        x1, y1 = max(0,x1), max(0,y1)
        x2, y2 = min(img_w, x2), min(img_h, y2)
        kx2, ky2 = kx1 + (x2 - x1), ky1 + (y2 - y1)
        dm[y1:y2, x1:x2] += KERNEL[ky1:ky2, kx1:kx2]
    return dm

class ShanghaiDataset(Dataset):
    """
    Expects:
      {part_A|part_B}/{train_data|test_data}/images/IMG_123.jpg
      {part_A|part_B}/{train_data|test_data}/ground-truth/GT_IMG_123.mat
    """
    def __init__(self, root, split='train', resize=512):
        split_folder = 'train_data' if split=='train' else 'test_data'
        self.imgs = sorted(glob.glob(os.path.join(root, split_folder, 'images', 'IMG_*.jpg')))
        # Map IMG_###.jpg -> GT_IMG_###.mat
        self.gts  = [os.path.join(root, split_folder, 'ground-truth',
                                  'GT_' + os.path.basename(p).replace('.jpg','.mat'))
                     for p in self.imgs]
        self.resize = resize

    def __len__(self): return len(self.imgs)

    def __getitem__(self, idx):
        img = cv2.imread(self.imgs[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        h, w = img.shape[:2]

        mat = sio.loadmat(self.gts[idx])
        # ShanghaiTech often stores points under image_info[0,0][0,0][0] or 'image_info' → 'location'
        pts = None
        if 'annPoints' in mat:
            pts = mat['annPoints']
        else:
            try:
                # Part A/B standard
                pts = mat['image_info'][0,0][0,0][0]
            except Exception:
                # some variants use key 'location'
                if 'image_info' in mat and 'location' in mat['image_info'][0,0].dtype.fields:
                    pts = mat['image_info'][0,0]['location'][0,0]
        if pts is None:
            raise RuntimeError(f"Unknown GT format for: {self.gts[idx]}")
        pts = np.array(pts, dtype=np.float32)
        if pts.ndim == 1:
            pts = pts.reshape(-1,2)

        # clamp
        pts[:,0] = np.clip(pts[:,0], 0, w-1)
        pts[:,1] = np.clip(pts[:,1], 0, h-1)

        # resize longest side to 512
        scale = self.resize / max(h, w)
        nh, nw = int(h*scale), int(w*scale)
        img_rs = cv2.resize(img, (nw, nh), interpolation=cv2.INTER_AREA)
        pts_sc = pts * scale

        dm = make_density_map(nh, nw, pts_sc)

        img_t = torch.from_numpy(img_rs.transpose(2,0,1)).float()/255.0
        dm_t  = torch.from_numpy(dm).unsqueeze(0).float()
        return img_t, dm_t
```

```python
import torch, torch.nn as nn, torch.nn.functional as F

class TinyCSRNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.frontend = nn.Sequential(
            nn.Conv2d(3,16,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(16,32,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32,64,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.backend = nn.Sequential(
            nn.Conv2d(64,64,3,padding=2,dilation=2), nn.ReLU(),
            nn.Conv2d(64,64,3,padding=2,dilation=2), nn.ReLU(),
            nn.Conv2d(64,1,1),
        )
    def forward(self,x):
        x = self.frontend(x)
        x = self.backend(x)
        return torch.relu(x)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = TinyCSRNet().to(device)
```

```python
class ShanghaiDataset(Dataset):
    def __init__(self, root, split='train', resize=512):
        split_folder = 'train_data' if split=='train' else 'test_data'
        self.imgs = sorted(glob.glob(os.path.join(root, split_folder, 'images', 'IMG_*.jpg')))
        self.gts  = [os.path.join(root, split_folder, 'ground-truth',
                                  'GT_' + os.path.basename(p).replace('.jpg','.mat'))
                     for p in self.imgs]
        self.resize = resize  # target square side (e.g., 512)

    def __len__(self): return len(self.imgs)

    def __getitem__(self, idx):
        import numpy as np, cv2, scipy.io as sio, torch

        img = cv2.imread(self.imgs[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        h, w = img.shape[:2]

        # --- load points robustly
        mat = sio.loadmat(self.gts[idx])
        pts = None
        if 'annPoints' in mat:
            pts = mat['annPoints']
        else:
            try:
                pts = mat['image_info'][0,0][0,0][0]
            except Exception:
                if 'image_info' in mat and 'location' in mat['image_info'][0,0].dtype.fields:
                    pts = mat['image_info'][0,0]['location'][0,0]
        if pts is None:
            raise RuntimeError(f"Unknown GT format for: {self.gts[idx]}")
        pts = np.array(pts, dtype=np.float32).reshape(-1,2)

        # --- fixed 512x512 (or self.resize x self.resize)
        target = self.resize
        sx = target / w
        sy = target / h
        img_rs = cv2.resize(img, (target, target), interpolation=cv2.INTER_AREA)

        # scale points per-axis
        pts_sc = np.empty_like(pts)
        pts_sc[:,0] = np.clip(pts[:,0] * sx, 0, target-1)
        pts_sc[:,1] = np.clip(pts[:,1] * sy, 0, target-1)

        # density map at fixed size
        dm = make_density_map(target, target, pts_sc)

        img_t = torch.from_numpy(img_rs.transpose(2,0,1)).float()/255.0
        dm_t  = torch.from_numpy(dm).unsqueeze(0).float()
        return img_t, dm_t
```

```python
import torch
import torch.nn.functional as F

def pad_collate(batch):
    """
    batch: list of (img[C,H,W], dm[1,H,W])
    Pads each to (C, H_max, W_max) and (1, H_max, W_max), then stacks.
    """
    imgs, dms = zip(*batch)
    H_max = max(x.shape[1] for x in imgs)
    W_max = max(x.shape[2] for x in imgs)

    imgs_p, dms_p = [], []
    for x, y in zip(imgs, dms):
        pad_h = H_max - x.shape[1]
        pad_w = W_max - x.shape[2]
        # F.pad uses (W_left, W_right, H_top, H_bottom)
        x_p = F.pad(x, (0, pad_w, 0, pad_h))
        y_p = F.pad(y, (0, pad_w, 0, pad_h))
        imgs_p.append(x_p)
        dms_p.append(y_p)

    return torch.stack(imgs_p, dim=0), torch.stack(dms_p, dim=0)
```

```python
import torch.nn as nn
import torch.optim as optim

# (Re)define model if needed, then:
def init_weights(m):
    if isinstance(m, (nn.Conv2d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity='relu')
        if m.bias is not None:
            nn.init.zeros_(m.bias)

model.apply(init_weights)

optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)  # LR up, add slight WD
```

```python
import torch.nn.functional as F
import torch

COUNT_W = 5e-4  # try 1e-4 to 1e-3 if needed
CLIP_NORM = 5.0

def train_epoch():
    model.train()
    tot = 0.0
    for img, dm in train_loader:
        img, dm = img.to(device), dm.to(device)

        # normalize to ~[-1, 1]
        img = (img - 0.5) / 0.5

        pred = model(img)
        if pred.shape[-2:] != dm.shape[-2:]:
            pred = F.interpolate(pred, size=dm.shape[-2:], mode='bilinear', align_corners=False)

        map_loss = F.mse_loss(pred, dm)
        count_loss = ((pred.sum(dim=(1,2,3)) - dm.sum(dim=(1,2,3)))**2).mean()
        loss = map_loss + COUNT_W * count_loss

        optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), CLIP_NORM)
        optimizer.step()
        tot += float(loss.item())
    return tot / max(1, len(train_loader))

def evaluate(loader):
    model.eval()
    mae = mse = n = 0
    with torch.no_grad():
        for img, dm in loader:
            img, dm = img.to(device), dm.to(device)
            img = (img - 0.5) / 0.5
            pred = model(img)
            if pred.shape[-2:] != dm.shape[-2:]:
                pred = F.interpolate(pred, size=dm.shape[-2:], mode='bilinear', align_corners=False)
            gt = dm.sum(dim=(1,2,3))
            pr = pred.sum(dim=(1,2,3))
            err = (pr - gt).abs()
            mae += float(err.mean().item())
            mse += float((err**2).mean().item())
            n += 1
    return mae/n, (mse/n)**0.5
```

```python
# take a tiny slice of the training set
mini_idx = list(range(8))  # 8 samples
from torch.utils.data import Subset, DataLoader
mini_loader = DataLoader(Subset(train_ds, mini_idx), batch_size=4, shuffle=True,
                         num_workers=0, pin_memory=True, collate_fn=pad_collate)

def train_overfit(mini_loader, epochs=5):
    for e in range(epochs):
        model.train()
        tot=0
        for img, dm in mini_loader:
            img, dm = img.to(device), dm.to(device)
            img = (img - 0.5) / 0.5
            pred = model(img)
            if pred.shape[-2:] != dm.shape[-2:]:
                pred = F.interpolate(pred, size=dm.shape[-2:], mode='bilinear', align_corners=False)
            map_loss = F.mse_loss(pred, dm)
            count_loss = ((pred.sum(dim=(1,2,3)) - dm.sum(dim=(1,2,3)))**2).mean()
            loss = map_loss + COUNT_W * count_loss
            optimizer.zero_grad(); loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), CLIP_NORM)
            optimizer.step()
            tot += float(loss.item())
        # eval the same mini set
        mae=0; n=0
        with torch.no_grad():
            for img, dm in mini_loader:
                img, dm = img.to(device), dm.to(device)
                img = (img - 0.5) / 0.5
                pred = model(img)
                if pred.shape[-2:] != dm.shape[-2:]:
                    pred = F.interpolate(pred, size=dm.shape[-2:], mode='bilinear', align_corners=False)
                gt = dm.sum(dim=(1,2,3)); pr = pred.sum(dim=(1,2,3))
                mae += float((pr-gt).abs().mean().item()); n += 1
        print(f"[Overfit] Epoch {e+1} loss={tot/len(mini_loader):.6f} | mini-MAE={mae/n:.2f}")

train_overfit(mini_loader, epochs=5)
```

```python
for epoch in range(10):
    tr = train_epoch()
    mae, rmse = evaluate(test_loader)
    print(f"Epoch {epoch+1}: train_loss={tr:.6f} | MAE={mae:.2f} | RMSE={rmse:.2f}")
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=1)
best_mae = float('inf'); history = []
for epoch in range(25):  # run ~25 more
    tr = train_epoch()
    mae, rmse = evaluate(test_loader)
    history.append((epoch+6, tr, mae, rmse))  # continue numbering if you want
    print(f"Epoch {epoch+6}: train_loss={tr:.6f} | MAE={mae:.1f} | RMSE={rmse:.1f}")
    scheduler.step(mae)
    if mae < best_mae:
        best_mae = mae
        torch.save(model.state_dict(), '/content/best_tinycsrnet.pth')
        print("✓ Saved new best with MAE", best_mae)
```

```python
df['t'] = pd.to_timedelta(df['time_s'], unit='s')
df = df.set_index('t')

# rolling 30-second average
df['roll_30s'] = df['people'].rolling('30s', min_periods=1).mean()

# robust threshold: median + 2 * MAD (manual MAD)
mad = (df['roll_30s'] - df['roll_30s'].median()).abs().mean()
thr = df['roll_30s'].median() + 2 * mad

df['over_threshold'] = df['roll_30s'] > thr
```
